```python
import numpy as np
import pandas as pd
data=pd.read_csv('/IRIS.csv')
```

```python
print(data.shape)
data.drop_duplicates(inplace=True)
print(data.shape)
print(f"After updation duplicate={data.duplicated().sum()}")
```

**Reasoning**:
Drop the duplicate rows from the DataFrame and verify the removal.



```python
display(data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=data)
plt.title('Petal Length vs. Petal Width by Species')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='petal_length', kde=True)
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='sepal_width', y='species', data=data)
plt.title('Box plot of Sepal Width by Species')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Species')
plt.show()
```
